```python
#Reading in GDP data
gdp_data = pd.read_csv("../Resources/GDP.csv")
#Renaming index for uniformity
gdp_data.rename(columns={"DATE":"Date"}, inplace=True)
gdp_grp = gdp_data.groupby('Date').sum()
#Displaying plot
gdp_grp.hvplot.bar(rot=90)
```

```python
cad_volume = pd.read_csv("../Resources/cad_volume.csv")
usd_volume = pd.read_csv("../Resources/usd_volume.csv")
mxn_volume = pd.read_csv("../Resources/mxn_volume.csv")

cad_volume.rename(columns={"Label":"Date"}, inplace=True)
usd_volume.rename(columns={"Label":"Date"}, inplace=True)
mxn_volume.rename(columns={"Label":"Date"}, inplace=True)
```

```python
#Dashboard construction

#Individual Currency Plots
def Currency_to_USD():
    cad_plot = currency_data["CAD to USD"].hvplot()
    mxn_plot = currency_data["MXN to USD"].hvplot()
    btc_plot = currency_data["BTC to USD"].hvplot()
    currency_row = pn.Row(cad_plot, mxn_plot, btc_plot)
    return currency_row

#Total Data Plot
def Comparison_plot():
    total_plot = currency_data.hvplot()
    return total_plot

#US GDP Data
def GDP_bplot():
    gdp_plot = gdp_grp.hvplot.bar(rot=90)
    return gdp_plot

#Individual Volume Plots
def Volume_plots():
    cad_vplot = cad_volume.hvplot(label="Weekly LocalBitcoins Volume, Canadian Dollar")
    usd_vplot = usd_volume.hvplot(label="Weekly LocalBitcoins Volume, US Dollar")
    mxn_vplot = mxn_volume.hvplot(label="Weekly LocalBitcoins Volume, Mexican Peso") 
    volume_row = pn.Row(cad_vplot, usd_vplot, mxn_vplot)
    return volume_row
```

```python
#Dashboard columns
currency_column = pn.Column(
    "## Value of Canadian Dollar, Mexican Peso and Bitcoin to US Dollar", Currency_to_USD()
)
total_column = pn.Column(
    "## Comparison of the Canadian Dollar, Mexican Peso and Bitcoin", Comparison_plot()
)
gdp_column = pn.Column(
    "## GDP of the United States", GDP_bplot()
)
volume_column = pn.Column(
    "## Volume of Bitcoin traded on LocalBitcoins by local market", Volume_plots()
)
#Creating Dashboard
currency_dashboard = pn.Tabs(
    ("Currency", currency_column),
    ("Comparison", total_column),
    ("GDP", gdp_column),
    ("Volume", volume_column)
)
```
